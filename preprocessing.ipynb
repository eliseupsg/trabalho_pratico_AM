```python
'''
Universidade Federal Fluminense
Programa de Pós-Graduação em Computação
Aprendizado de Máquina (EAD) - 2020.1
Autor: Eliseu Paz e Silva de Guimarães
Trabalho Prático
Descrição: este script tem o objetivo de realizar o pré-processamento dos dados utilizados no trabalho prático da disciplina
de AM (EAD)
São extraídos do arquivo inicial apenas os ratings (estrelas) e os textos dos reviews
São selecionadas 15.000 instâncias assim distribuídas:
1 estrela: 3.000 instâncias
2 estrelas: 3.000 instâncias
4 estrelas: 3.000 instâncias
5 estrelas: 3.000 instâncias
As instâncias são pré-processadas e salvas em um arquivo CSV
'''
import pandas as pd
import re
from nltk.tokenize import word_tokenize 

raw_file = open("all.review", "r", encoding="ansi")
get_rating = False
get_review_text = False
review_text = ""
rating = ""
ratings_qtt = [3000, 3000, 0, 3000, 3000]
df = pd.DataFrame(columns=['text', 'class'])
reviews_texts = []
classes = []

print("Lendo o arquivo inicial...")
for line in raw_file:
    
    #Verifica o fechamento do texto do review
    if "</review_text>" in line:
        get_review_text = False
    if get_review_text == True:
        review_text += line.rstrip()
    
    #Verifica se o rating está aberto
    if get_rating==True:
        rating = re.sub(".0\n", "", line.rstrip())
        get_rating = False
    
    #Verifica se o review_text ou o rating devem ser abertos
    if "<review_text>" in line:
        get_review_text = True
    if "<rating>" in line:
        get_rating = True
    
    #Verifica o fechamento do review
    if "</review>" in line:
        #Checa se o review tem uma quantidade de estrelas ainda disponível
        stars = int(float(rating))
        if ratings_qtt[stars-1]>0:
            #Remove as URLs do review_text
            review_text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+',"",
                    review_text, flags=re.MULTILINE)
            #Remove sinais de pontuação e caracteres especiais
            review_text = re.sub(r'[.,;:()%#$"\']',"", review_text, flags=re.MULTILINE)
            #Remove números
            review_text = re.sub(r'[0-9]',"", review_text, flags=re.MULTILINE)
            #Remove tabulações
            review_text = re.sub(r'\t'," ", review_text, flags=re.MULTILINE)
            #Tokeniza
            review_tokens = word_tokenize(review_text)
            #Junta os tokens
            review_text = ' '.join(review_tokens)
            
            reviews_texts.append(review_text)
            if stars>3:
                classes.append(1)
            else:
                classes.append(0)
 
            ratings_qtt[stars-1] = ratings_qtt[stars-1] - 1
            
            if (ratings_qtt[stars-1]%500)==0:
                print("Faltam "+str(ratings_qtt[stars-1])+" instâncias com "+str(stars)+" estrelas.")
        review_text = ""
        rating = ""
        
raw_file.close()
df['text'] = reviews_texts
df['class'] = classes
df.to_csv("pp_data.csv", index=False, header=False)
print("Arquivo pré-processado salvo.")

display(df)
```
